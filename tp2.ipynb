## Ejercicio 3 - Ajustes no polinomiales

```python
#Cargamos los datos
xi, yi = np.load("./datos/ejercicio_3_i.npy")
xii, yii = np.load("./datos/ejercicio_3_ii.npy")
print(xi, yi)
print(xii, yii)
```

a) Para los conjuntos de datos (3.i) y (3.ii) por separado, proponer una forma funcional adecuada y hallar la curva de la forma propuesta que mejor aproxima los datos en el sentido de cuadrados mínimos.

```python
#arr = np.array([xi, yi])
#np.fft.fft(arr)


import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Supongamos que tienes datos xi e yi
xi = np.array([1, 2, 3, 4, 5])
yi = np.array([2, 3, 5, 7, 11])

# Aplicar la FFT a los datos
fft_xi = np.fft.fft(xi)
fft_yi = np.fft.fft(yi)

# Definir una función que representa el ajuste basado en la transformada de Fourier
def fourier_fit(x, *params):
    n = len(params) // 2  # Obtener la mitad de los parámetros (amplitudes y fases)
    result = np.zeros_like(x, dtype=np.complex128)
    for i in range(n):
        result += params[i] * np.exp(1j * i * x + params[i + n])
    return np.real(result)

# Unir los datos de xi e yi para realizar un ajuste conjunto
x_combined = np.concatenate((xi, yi))
fft_combined = np.concatenate((np.abs(fft_xi), np.abs(fft_yi)))

# Ajustar la función fourier_fit a los datos
params, _ = curve_fit(fourier_fit, x_combined, fft_combined)

# Generar valores para la curva ajustada
x_vals = np.linspace(x_combined[0], x_combined[-1], 100)
y_vals = fourier_fit(x_vals, *params)

# Visualizar los datos y la curva ajustada
plt.scatter(xi, yi, label='Datos')
plt.plot(x_vals, y_vals, 'r', label='Curva ajustada')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.legend()
plt.title('Ajuste con Transformada de Fourier (mínimos parámetros)')
plt.show()

```

b) Graficar en dos figuras distintas los conjuntos de datos y las curvas ajustadas en cada caso.

```python
plt.scatter(xi, yi, label = "obs_i")
plt.scatter(xii, yii, label = "obs_ii")
```
